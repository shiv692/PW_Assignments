Q 1. What is the difference between a function and a method in Python?
  - In Python, a function is a block of reusable code that performs a specific task and is defined using the def keyword. A method is similar to a function but is associated with an object and is called on that object (it belongs to a class). Simply put, all methods are functions, but not all functions are methods. Functions stand alone, while methods are tied to objects
    - Example:
len() is a function used like len("hello"), but "hello".upper() is a method that works on string objects.



Q 2. Explain the concept of function arguments and parameters in Python
  - Parameters are the variable names listed in a function definition, while arguments are the actual values passed when calling the function. Parameters act as placeholders, and arguments provide real data for the function to work with. Python supports different types of arguments: positional, keyword, default, and variable-length.
    
    Example:
            def greet(name):   # 'name' is a parameter
            print("Hello", name)
            
            greet("Pranav")    # "Pranav" is the argument

Q 4. What is the purpose of the return statement in a Python function?

Q 5. What are iterators in Python and how do they differ from iterables?
  - An iterator is an object that allows you to go through all the elements in a sequence one by one using the next() function. It remembers its position while iterating. An iterable is any object that can return an iterator, like a list, tuple, or string. The main difference is: iterables can be looped over, while iterators actually perform the iteration.
    
    Example:

Q 6. Explain the concept of generators in Python and how they are defined
  - Generators are a type of iterator that are used to produce a sequence of values lazily—meaning one at a time and only when needed. They are defined using functions with the yield keyword instead of return. Generators are memory-efficient and useful for large datasets.
  
    Example:
            def count_up_to(n):
            i = 1
             while i <= n:
              yield i
               i += 1
               
               for num in count_up_to(3):
                   print(num)    # prints 1, 2, 3 one by one


Q 7. What are the advantages of using generators over regular functions
  - Generators are better than regular functions when we want to generate large sequences of data without using much memory. Instead of returning all values at once (like a list), they yield one value at a time. This makes them faster and more memory-efficient, especially with loops or data streams.
  
    Example:
            def generate_numbers():
             for i in range(1000000):
              yield i

      This generator saves memory compared to a list with a million numbers.


Q 8. What is a lambda function in Python and when is it typically used?
  - A lambda function is a small anonymous function (without a name). It can take any number of arguments but has only one expression. It’s mostly used when we need a simple function for a short time — often inside functions like map(), filter(), or sorted().
  
    Example:
              square = lambda x: x * x

Q 9. Explain the purpose and usage of the map() function in Python.
  - The map() function is used to apply a function to each item in a list (or any iterable) and return a new iterable (like a map object). It’s useful when we want to transform every item of a sequence in the same way.
  
    Example:
            nums = [1, 2, 3, 4]
              squares = list(map(lambda x: x*x, nums))  
                print(squares)  # Output: [1, 4, 9, 16]


Q 10. What is the difference between map(), reduce(), and filter() functions in Python?
  - map() is used to apply a function to every item in a list (or any iterable) and returns a new list of results.
  - filter() is used to select items from a list based on a condition (only the items for which the function returns True).
  - reduce() is used to perform a calculation across all items and reduce them to a single value (like sum, product, etc.).

   reduce() is part of the functools module and needs to be imported.
   
    Example:
            from functools import reduce
            nums = [1, 2, 3, 4, 5]
            # map → squares of each number
            squares = list(map(lambda x: x*x, nums))
            # filter → keep even numbers
            evens = list(filter(lambda x: x % 2 == 0, nums))
            # reduce → multiply all numbers
            product = reduce(lambda x, y: x * y, nums)

```python
# Question 1: Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even_numbers(numbers):
    total = 0

    for num in numbers:
        if num % 2 == 0:
            total += num

    return total

# Example usage
sample_list = [1, 2, 3, 4, 5, 6]
print("Sum of even numbers:", sum_of_even_numbers(sample_list))
```

```python
# Question 3: Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def square_numbers(numbers):
    squared_list = []

    for num in numbers:
        squared_list.append(num ** 2)

    return squared_list

# Example usage
sample_list = [1, 2, 3, 4, 5]
print("Squared numbers:", square_numbers(sample_list))
```

```python
# Question 4: Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n <= 1:
        return False

    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False

    return True

# Example usage
for num in range(1, 201):
    if is_prime(num):
        print(num, "is a prime number")
```

```python
# Question 5: Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    """Iterator that returns Fibonacci numbers up to `max_terms`."""

    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration

        current = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return current

# Example usage
fib = FibonacciIterator(10)
for num in fib:
    print(num, end=" ")

```

```python
# Question 6: Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(max_exponent):
    """
    Yields 2**0, 2**1, …, up to 2**max_exponent (inclusive).
    """
    exponent = 0
    while exponent <= max_exponent:
        yield 2 ** exponent
        exponent += 1

# Example usage
for power in powers_of_two(5):
    print(power)

```

```python
# Question 7: Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_lines(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            yield line

# Create a dummy file for testing
with open("sample.txt", "w", encoding="utf-8") as f:
    f.write("This is line 1.\n")
    f.write("This is line 2.\n")
    f.write("This is line 3.\n")


# Example usage to print the lines (test)
for line in read_file_lines("sample.txt"):
    print(line.strip())
```

```python
# Question 8: Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

def sort_by_second_element(tuples_list):
    return sorted(tuples_list, key=lambda x: x[1])

# Example usage
data = [(1, 3), (2, 1), (4, 2)]
print(sort_by_second_element(data))
```

```python
# Question 9: Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit.

def convert_to_fahrenheit(celsius_list):
    return list(map(lambda c: (c * 9/5) + 32, celsius_list))

# Example usage
temps_celsius = [0, 20, 30, 37]
print(convert_to_fahrenheit(temps_celsius))
```

```python
# Question 10: Create a Python program that uses filter() to remove all the vowels from a given string.

def remove_vowels(input_string):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, input_string))

# Example usage
text = "Hello World"
print(remove_vowels(text))
```
